```python
alltags = {}
tagpairs = {}
def crawlTweets(query, level = 3):
    print "CURRENT QUERY: %s, level = %d"%(query, level)
    level -= 1
    if level > 0:
        tags, users = {}, {}
        tweets = t.search.tweets(q=query, count=100)
        for tweet in tweets['statuses']:
            words = tweet['text'].split()
            tag_in_text = []
            for i, word in enumerate(words):
                if word.endswith(tuple(':,.-_')):
                    words[i] = words[i][:-1]
            for word in words:
                if word.startswith("#"):
                    if word not in tag_in_text:
                        tag_in_text.append(word)
            if len(tag_in_text) > 1:
                for t1 in tag_in_text:
                    for t2 in tag_in_text:
                        if t2 != t1:
                            tagpairs.setdefault(t1, {})
                            tagpairs[t1].setdefault(t2, 0)
                            tagpairs[t1][t2] += 1
            for word in words:
                if word.startswith("#"):
                    tags.setdefault(word, 0)
                    tags[word] += 1
                if word.startswith("@"):
                    users.setdefault(word, 0)
                    users[word] += 1
        for tag in tags:
            alltags.setdefault(tag, 0)
            alltags[tag] += tags[tag]
        for tag in tags:
            crawlTweets(tag, level)
crawlTweets('obama', 3)
        
```
